```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merged_df = pd.merge(clinical_df, mouse_df, on="Mouse ID")

# Display the data table for preview
merged_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_df = merged_df.groupby(["Drug", "Timepoint"])
avg_Tumor_Volume = grouped_df.mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
avg_Tumor_Volume_df = pd.DataFrame(avg_Tumor_Volume)

# Preview DataFrame
avg_Tumor_Volume_df = avg_Tumor_Volume_df.reset_index()
avg_Tumor_Volume_df.head()
```

```python
#preview the grouped dataframe to go back and reference
grouped_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
Standard_Error_of_Tumor_Volume = grouped_df.sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
Standard_Error_of_Tumor_Volume_df = pd.DataFrame(Standard_Error_of_Tumor_Volume)

# Preview DataFrame
Standard_Error_of_Tumor_Volume_df = Standard_Error_of_Tumor_Volume_df.reset_index()
Standard_Error_of_Tumor_Volume_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
avg_Tumor_Volume_munged_df = avg_Tumor_Volume_df.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")
# Preview that Reformatting worked
avg_Tumor_Volume_munged_df
```

```python
# Generate the Plot (with Error Bars)
#Creating a scatter plot that shows how the tumor volume changes over time for each treatment.
#Create a fig in order to share the same y-axis for each drug
fig, ax = plt.subplots(figsize=(12,8))

#Set the x-axis as the values of the index or Timepoint
x_axis = avg_Tumor_Volume_munged_df.index.values

#Create lists of colors
colors = ["C0", "C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9"]

#Create a loop to go through each drug in each column
x = 0
for drug in avg_Tumor_Volume_munged_df.columns:
    y_axis = avg_Tumor_Volume_munged_df[drug].values
    #Add error bars to each data point on the scatter plot
    #Use the standard error data frame to find the value of each error bar at each data point
    standard_errors = Standard_Error_of_Tumor_Volume_df[Standard_Error_of_Tumor_Volume_df.Drug == drug]["Tumor Volume (mm3)"].values
    ax.errorbar(x_axis, y_axis, standard_errors, fmt="o", label=drug, color=colors[x], linestyle="solid")
    x = x + 1


#Add titles and labels to the plot to look nice and professional
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc="best")
plt.grid()

# Save the Figure
plt.savefig("Tumor_Response_to_Treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
avg_Metastatic_Response = grouped_df.mean()["Metastatic Sites"]
# Convert to DataFrame
avg_Metastatic_Response_df = pd.DataFrame(avg_Metastatic_Response)
# Preview DataFrame
avg_Metastatic_Response_df = avg_Metastatic_Response_df.reset_index()
avg_Metastatic_Response_df
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
#Reference the previous standard error code
Standard_Error_of_Metastatic = grouped_df.sem()["Metastatic Sites"]

# Convert to DataFrame
Standard_Error_of_Metastatic_df = pd.DataFrame(Standard_Error_of_Metastatic)

# Preview DataFrame
Standard_Error_of_Metastatic_df = Standard_Error_of_Metastatic_df.reset_index()
Standard_Error_of_Metastatic_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
#Reference the previous data munging
avg_Metastatic_Response_Munged_df = avg_Metastatic_Response_df.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")

# Preview that Reformatting worked
avg_Metastatic_Response_Munged_df
```

```python
# Generate the Plot (with Error Bars)
#Reference the previous plot
fig, ax = plt.subplots(figsize=(12,8))

x_axis = avg_Metastatic_Response_Munged_df.index.values

colors = ["C0", "C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9"]

x = 0
for drug in avg_Metastatic_Response_Munged_df.columns:
    y_axis = avg_Metastatic_Response_Munged_df[drug].values
    standard_errors = Standard_Error_of_Metastatic_df[Standard_Error_of_Metastatic_df.Drug == drug]["Metastatic Sites"].values
    ax.errorbar(x_axis, y_axis, standard_errors, fmt="o", label=drug, color=colors[x], linestyle="solid")
    x = x + 1


#Add titles and labels to the plot to look nice and professional
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Metastatic Sites (mm3)")
plt.legend(loc="best")
plt.grid()

# Save the Figure
plt.savefig("Metastatic_Spread_During_Treatment.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
Survival_Rates = grouped_df.count()["Mouse ID"]
# Convert to DataFrame
Survival_Rates_df = pd.DataFrame(Survival_Rates)
# Preview DataFrame
Survival_Rates_df.rename(columns={"Mouse ID": "Mouse Count"})
Survival_Rates_df = Survival_Rates_df.reset_index()
Survival_Rates_df
#Renaming the columns is not giving me an error but it is not changing the column name
```

```python
# Minor Data Munging to Re-Format the Data Frames
Survival_Rates_Munged_df = Survival_Rates_df.pivot(index="Timepoint", columns="Drug", values="Mouse ID")
# Preview the Data Frame
Survival_Rates_Munged_df
```

```python
# Generate the Plot (Accounting for percentages)
fig, ax = plt.subplots(figsize=(12,8))

x_axis = Survival_Rates_Munged_df.index.values

colors = ["C0", "C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9"]

x = 0
for drug in Survival_Rates_Munged_df.columns:
    #get number of mice to get a percentage
    number_of_mice = Survival_Rates_Munged_df[drug].values[0]
    y_axis = Survival_Rates_Munged_df[drug].values / number_of_mice * 100
    plt.plot(x_axis, y_axis, label=drug, marker="o", color=colors[x])
    x = x + 1


#Add titles and labels to the plot to look nice and professional
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend(loc="best")
plt.grid()
# Save the Figure
plt.savefig("Survival_During_Treatment.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
avg_Tumor_df = avg_Tumor_Volume_df.drop(["Timepoint"], axis=1)
Drug_grouped_df = avg_Tumor_df.groupby(["Drug"])

Percent_Changes = (Drug_grouped_df["Tumor Volume (mm3)"].last() - Drug_grouped_df["Tumor Volume (mm3)"].first()) / Drug_grouped_df["Tumor Volume (mm3)"].first() * 100
# Display the data to confirm
Percent_Changes
```

```python
# Store all Relevant Percent Changes into a Tuple
Percent_Changes = Percent_Changes.sort_values(ascending=True)

# Splice the data between passing and failing drugs
Failed_Drug = Percent_Changes[Percent_Changes < 0].sort_values()
Pass_Drug = Percent_Changes.copy()
Pass_Drug.loc[Pass_Drug.values < 0] = 0
Pass_Drug = Pass_Drug.sort_values()

x_axis = np.arange(len(Percent_Changes))
tick_locations = [value+0.5 for value in x_axis]

plt.figure(figsize=(14,4))

ax = Failed_Drug.plot(kind='bar', color='g', width=1)
Pass_Drug.plot(kind='bar', color='r', width=1, ax=ax)
plt.xticks(tick_locations, Percent_Changes.index)

plt.title("Tumor Change Over 45 Days Treatment")
plt.ylabel("% Tumor Volume Change")
plt.xlim(-0.7,len(Percent_Changes)-0.3)
plt.ylim(-30,70)
plt.grid()

rects = ax.patches
for rect in rects:
    # Get X and Y placement of label from rect.
    y_value = rect.get_height()
    if y_value < 0.0:
        x_value = rect.get_x() + rect.get_width() / 2
        percentY = str(int(y_value)) + '%'
        plt.text(x_value-0.1, -5, percentY, color='w')
    elif y_value > 0.0:
        x_value = rect.get_x() + rect.get_width() / 2
        percentY = str(int(y_value)) + '%'
        plt.text(x_value-0.1, 5, percentY, color='w')

# Save the Figure
plt.savefig("Tumor_Change_Over_45_Days.png")

# Show the Figure
fig.show()
```
